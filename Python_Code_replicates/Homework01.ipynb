```python
import matplotlib.pyplot as plt
```

```python
layers = [x for x in range(1, 101)]
```

```python
thickness = 63700
radius =[6370000]
initial = 6370000
for x in range(1, len(layers)):
    initial = initial - thickness
    radius.append(initial)
```

```python
density = [8811-(0.000865*x) for x in radius][:47]
to_add = [11000 for x in range(53)]
density.extend(to_add)
```

```python
inertia = [(8/3) * 3.142 * density[x] * (radius[x]**4) *thickness for x in range(100)]
```

```python
mass = [4*3.142* radius[x]**2 * thickness * density[x] for x in range(100)]
```

```python
plt.plot(radius, density, '.')
plt.xlabel("Radius (meter)")
plt.ylabel("Density(kg per m^3)")
plt.title("Density as a function of Radius", fontdict={'fontsize':15, 'fontweight': 'black', 'color':'blue'}, pad = 10);
```

```python
plt.plot(radius, mass, '.')
plt.xlabel("Radius (meter)")
plt.ylabel("Mass (kg)")
plt.title("Mass as a function of Radius", fontdict={'fontsize':15, 'fontweight': 'black', 'color':'red'}, pad = 10);
```

```python
plt.plot(radius, inertia, '.')
plt.xlabel("Radius (meter)")
plt.ylabel("Inertia (kg per m^2)")
plt.title("Density as a function of Radius", fontdict={'fontsize':15, 'fontweight': 'black', 'color':'gray'}, pad = 10);
```
